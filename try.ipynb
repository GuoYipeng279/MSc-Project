```python
# Example scatter_plots list (replace this with your own data)
scatter_plots = [
    np.random.rand(50, 2),  # Scatter plot 1
    np.random.rand(50, 2),  # Scatter plot 2
    np.random.rand(50, 2),  # Scatter plot 3
    # Add more scatter plots here...
]
```

```python
fig, ax = plt.subplots()
sc = ax.scatter([], [])  # Empty scatter plot
```

```python
def init():
    sc.set_offsets([])  # Clear the scatter plot
    return (sc,)
def update(frame):
    data = scatter_plots[frame]  # Get the data for the current frame
    sc.set_offsets(data)  # Set the new data for the scatter plot
    return (sc,)
```

```python
ani = FuncAnimation(fig, update, frames=len(scatter_plots), init_func=init, blit=True)
```

```python
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

# Sample data for scatter plots
x = [1, 2, 3, 4, 5]
y = [3, 1, 4, 2, 5]

# Function to update the scatter plot in each frame of the animation
def update(frame):
    plt.cla()  # Clear the previous plot
    plt.scatter(x[frame:], y[frame:], c='b', marker='o')  # Create the scatter plot
    plt.xlim(0, 6)
    plt.ylim(0, 6)
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.title('Scatter Plot Animation')
    plt.plot([0, 6], [0, 6], 'r--')  # Add a line y=x to the plot

# Create the animation
fig, ax = plt.subplots()
ani = FuncAnimation(fig, update, frames=10, interval=500)

# Save the animation to a video file
ani.save('scatter_animation.mp4', writer='ffmpeg')

plt.show()
```
